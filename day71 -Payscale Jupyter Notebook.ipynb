# Updated Payscale Data

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

base_url='https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors'

major_data_list = []
headings = ['Rank', 'Major', 'Degree Type', 'Early Career Pay', 'Mid-Career Pay', '% High Meaning']
for i in range (1, 35):
  page_url = base_url
  if i > 1:
    page_url += f'/page/{i}'
  resp = requests.get(page_url)
  ps_page = resp.text
  soup = BeautifulSoup(ps_page, 'html.parser')
  majors = soup.tbody.find_all("tr", class_="data-table__row")
  for major in majors:
    cell_value_list = []
    cells = major.find_all("span", class_="data-table__value")
    for cell in cells:
      cell_value_list.append(str(cell.string))
    major_data_list.append(cell_value_list)

ps = pd.DataFrame(major_data_list, columns=headings)
ps["Early Career Pay"] = ps["Early Career Pay"].replace('[\$,\"]', '', regex=True).astype(int)
ps["Mid-Career Pay"] = ps["Mid-Career Pay"].replace('[\$,\"]', '', regex=True).astype(int)
ps["% High Meaning"] = ps["% High Meaning"].replace('%', '', regex=True).replace('-', 0, regex=True).astype(int)
ps.to_csv("payscale.csv", index=False)


```
